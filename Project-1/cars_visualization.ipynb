```python
# importing intial libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Motor Trend Car Road Tests

Description <br>
The data was extracted from the 1974 Motor Trend US magazine, and comprises fuel consumption and 10 aspects of automobile design and performance for 32<br> automobiles (1973–74 models).<br>

Format<br>
A data frame with 32 observations on 11 variables.<br>

1 - mpg: Miles/(US) gallon<br>

2 - cyl: Number of cylinders<br>

3 - disp: Displacement (cu.in.)<br>

4 - hp: Gross horsepower<br>

5 - drat: Rear axle ratio<br>

6 - wt: Weight (1000 lbs)<br>

7 - qsec: 1/4 mile time<br>

8 - vs: Engine shape (0 = v-shaped, 1 = straight)<br>

9 - am: Transmission (0 = automatic, 1 = manual)<br>

10 - gear: Number of forward gears<br>

11 - carb: Number of carburettors<br>

Source<br>
Henderson and Velleman (1981), Building multiple regression models interactively. Biometrics, 37, 391–411.

```python
# reading csv into dataframe
df = pd.read_csv("cars.csv")

# renaming column car to model
df.rename(columns={'Car':'model'}, inplace=True)

# Printing first few rows of the dataframe
df.head()
```

# BARPLOT

A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. <br>
Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
# barplot between no of cylinders and miles per gallon
plt.figure(figsize=(6,4))
sns.barplot(x = 'cyl', y = 'mpg', data=df, palette='Set1')
plt.xlabel('No of Cylinders')
plt.ylabel('Miles per(US) gallon')
plt.title('CYL VS MPG')
plt.show()
```

##### INFERENCE
We can infer from the plot that as the number of cylinders increases, the miles per gallon for any vehicle decreases.

```python
# plotting the barplot for transmission(am) vs mpg
plt.figure(figsize=(6,4))
sns.barplot(x = 'am', y = 'mpg', data=df, palette='Set2')
plt.xlabel('Automatic Transmission')
plt.ylabel('Miles per(US) gallon')
plt.title('AM VS MPG')
plt.show()
```

##### INFERENCE
We can infer from the plot that manual transmission(am = 1) is much more fuel efficient that automatic transmission(am = 0) for most of the cars.

```python
# barplot for gear vs mpg
plt.figure(figsize=(6,4))
sns.barplot(x = 'gear', y = 'mpg', data=df, palette='Set3')
plt.xlabel('No of Forward Gears')
plt.ylabel('Miles per(US) gallon')
plt.title('GEARS VS MPG')
plt.show()
```

##### INFERENCE

A 4 geared car is generally more efficient than 3 or 5.

```python
# barplot for carb vs mpg
plt.figure(figsize=(6,4))
sns.barplot(x = 'carb', y = 'mpg', data=df, palette='rocket')
plt.xlabel('No of Carburettors')
plt.ylabel('Miles per(US) gallon')
plt.title('CARB VS MPG')
plt.show()
```

##### INFERENCE
Cars with 1 carburettors are more efficient than those that have higher than them

```python
# barplot for gears vs horsepower
plt.figure(figsize=(6,4))
sns.barplot(x = 'gear', y = 'hp', data=df, palette='dark')
plt.xlabel('No of Forward Gear')
plt.ylabel('Gross Horsepower')
plt.title('Gear VS HP')
plt.show()
```

##### INFERENCE
Cars with 4 gears have lowest horsepower as compared to others.<br>
Cars with 5 gears have highest horsepower

# COUNTPLOT
Show the counts of observations in each categorical bin using bars.

```python
# horizontal countplot for gears
plt.figure(figsize=(6,4))
sns.countplot(y = 'gear', data=df, palette='muted')
plt.ylabel('GEARS')
plt.show()
```

##### INFERENCE
We have maximum number of cars having 3 gears followed by 4 and 5 gears

```python
# horizontal countplot for transmission
plt.figure(figsize=(6,4))
sns.countplot(y = 'am', data=df, palette='bright')
plt.ylabel('Transmission')
plt.show()
```

##### INFERENCE
More cars have automatic transmission than a manual one

```python
# horizontal countplot for carburettors
plt.figure(figsize=(6,4))
sns.countplot(y = 'carb', data=df, palette='hls')
plt.ylabel('Carburettors')
plt.show()
```

##### INFERENCE
Cars mostly have 2 or 4 carburettors while very rarely have 6 or 8 carburettors

```python
# horizontal countplot for vs
plt.figure(figsize=(6,4))
sns.countplot(y = 'vs', data=df, palette='husl')
plt.ylabel('ENGINE SHAPE')
plt.show()
```

##### INFERENCE
More cars have v-shaped engine(vs=0) than a straight one(vs=1)

```python
# horizontal countplot for cylinders
plt.figure(figsize=(6,4))
sns.countplot(y = 'cyl', data=df, palette='Paired')
plt.ylabel('No of Cylinders')
plt.show()
```

##### INFERENCE
Most cars have 8 cylinders while very less cars have 6 cylinders

```python
# countplot for cars having different gears grouped by cylinders
plt.figure(figsize=(6,4))
sns.countplot(x = 'gear', data=df, palette='rocket', hue='cyl')
plt.xlabel('GEARS')
plt.show()
```

##### INFERENCE
For cars having 3 gears, most of them have 8 cylinders<br>
For cars having 4 gears, majority of them have 4 cylinders<br>
For cars having 5 gears, there is not much difference in no of carburettors

# HEATMAPS
They show the correlation between various feature variables of the dataset. Top bar denotes positive correlation and vice versa.

```python
# heatmap for all features
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only = True), cbar=True, linewidths=0.5, linecolor='black')
plt.show()
```

##### INFERENCE
We can infer correlation by the color, ex, cyl and mpg have dark color, meaning negative correlation which can be inferred as greater no of cylinders, less the miles per gallon.

# SCATTERPLOT
A graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
# scatter plot for weight and qsec(1/4 mile time) labelled by the no of cylinders they have
plt.figure(figsize=(6,4))
sns.scatterplot(x = 'wt', y = 'qsec', data = df, hue = 'cyl', palette='dark')
plt.xlabel('Weight(1000 lbs)')
plt.ylabel('Q-sec(1/4 mile time)')
plt.title('Scatter PLOT btwn QSEC and Weight')
plt.show()
```

##### INFERENCE
Cars having 8 cylinders are typically heavier than others but still manage to be faster(less qsec) than others<br>
Cars with 4 cylinders are usually light in weight but are still slower(greater qsec) than others

```python
# scatter plot between MPG and HP
plt.figure(figsize=(6,4))
sns.scatterplot(x = 'mpg', y = 'hp', data = df, hue = 'cyl', palette='bright', size='am')
plt.xlabel('Miles per(US) Gallon')
plt.ylabel('Horsepower')
plt.title('Scatter PLOT btwn MPG and HP')
plt.show()
```

##### INFERENCE
Cars having 4 cylinders are mostly manual, having lower horsepower but are the most efficient(higher miles per gallon)<br>
Cars having 8 cylinders are mostly automatic, having higher horsepower but are very less efficient(lower miles per gallon)

```python
# scatter plot between disp and qsec
plt.figure(figsize=(6,4))
sns.scatterplot(x = 'disp', y = 'qsec', data = df, hue = 'cyl', palette='Set1', size='vs')
plt.xlabel('Displacement')
plt.ylabel('Q-sec')
plt.title('Scatter PLOT btwn DISPLACEMENT and Q-sec')
plt.show()
```

##### INFERENCE
Cars having 8 cylinders have v-shaped(vs=0) engines and have lower qsec inspite of having higher displacement
Cars having 4 cylinders have majorly straight(vs=1) engines and have higher qsec inspite of having lower displacement

# LINEAR REGRESSION PLOT
Shows a linear regression line accross features of the dataframe

```python
# linear regression plot of mpg vs hp
plt.figure(figsize=(6,4))
sns.lmplot(x = 'mpg', y = 'hp', data=df)
plt.xlabel("Miles Per(US) Gallon")
plt.ylabel("Horsepower")
plt.title("LINEAR REG PLOT btwn HP AND MPG")
plt.show()
```
